```python
## IMPORT LIBRARIES#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tracks = pd.read_csv('tracks.csv')
```

```python
tracks.shape ## row and columns of dataset#
```

```python
tracks.info() 
```

```python
tracks.isnull().sum() ##to check NA or Null values/ missing values#
```

```python
sorted_track =tracks.sort_values('popularity', ascending=True).head(10) ## for checking least popular song
sorted_track
```

```python
sorted_track1= tracks.sort_values('popularity', ascending=False).head(10) ## for checking top popular song
sorted_track1
```

```python
tracks.describe().transpose()
```

```python
pp_tracks = tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False).head(10)
pp_tracks
```

```python
##to create index for release date#
tracks.set_index('release_date', inplace= True)
tracks.index = pd.to_datetime(tracks.index)
tracks.head()
```

```python
tracks[['artists']].iloc[19] ## finding specific artist name#
```

```python
tracks['duration']=tracks['duration_ms'].apply(lambda x:round(x/1000)) ##convertin millisecond to second
tracks.drop('duration_ms',inplace=True,axis= 1) ##dropping duration ms as we have converted to second
```

```python
tracks.duration.head()
```

```python
tracks.columns
```

```python
##Correlation heat map#
corr_track = tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_track,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno', linewidths=1,linecolor='Black')
heatmap.set_title("Correlation heatmap btw variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation= 90)
```

```python
sample_df = tracks.sample(int(0.004*len(tracks)))
print(len(sample_df))
```

```python
## lets create regression plt btw loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness', x= 'energy', color='c')
plt.title('loudness vs energy correlation')
plt.show()
```

- There is high Correlation Between these two

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity', x= 'acousticness', color='b')
plt.title('popularity vs acousticness correlation')
plt.show()
```

- There is negative Correlation Between these two

```python
tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

```python
## lets see distribution of song in year from 1922 to 2020
sns.displot(years, discrete=True,aspect=3,height=5,kind='hist', color='Turquoise', edgecolor='black')
plt.title('No of songs per year')
plt.show()
```

```python
# year vs duration in Bar plot #
total_duration = tracks.duration
fig_dims=(20,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig = sns.barplot(x=years, y=total_duration, ax=ax, errwidth=False)
plt.title('years vs duration')
plt.xticks(rotation=90)
```

- From 1900 to 1935 the duration were less
- Post 1935 till 2010 almost there is consistency in duration
- Post 2010 it again started decreasing


## PROJECT 1 :SPOTIFY DATA ANALYSIS BY GENRE


```python
## IMPORT LIBRARIES#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tracks_genre = pd.read_csv('SpotifyFeatures.csv')
```

```python
tracks_genre.head()
```

```python
tracks_genre.shape
```

```python
##Ploting bar plot for duration of song wrt genre
plt.title('Duration of Song in different genre')
sns.color_palette('rocket',as_cmap=True)
sns.barplot(y='genre', x='duration_ms' , data=tracks_genre)
plt.xlabel('Duration in MS')
plt.ylabel('genre')

```

- In Graph we can see Classical and World Genre Duration 
- Next is Jazz and Electronic

```python
## Top 5 genre by popularity
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous = tracks_genre.sort_values('popularity', ascending=False).head(10) ## Taken 10 only because of repetitive genre in data
sns.barplot(y='genre', x='popularity', data=famous)
plt.title('Top 5 genre based on popularity')
plt.show()
```
